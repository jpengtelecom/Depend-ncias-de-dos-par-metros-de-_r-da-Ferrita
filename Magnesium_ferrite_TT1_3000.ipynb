```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
C = 3 * 10**8  # Velocidade da luz (m/s)
mi0 = 4 * np.pi * (10**(-7))  # Permeabilidade do vcuo
Epsilon_0 =  8.8541878176 * 10**(-12) # permissividade do vcuo  (F/m)
Gamma = 1.759 * (10**11)  # Gyromagnetic ratio (Ckg1)

# Caractersticas do Magnesium ferrite TT1-3000 - Apndice H do Livro do Pozar

# 4**10**(3) oersted (Oe) = 1 (A/m)
H0_values = [3200, 3400, 3600]  # Valores de H0 para diferentes simulaes (Oersteds)

Delta_H = 190  # Oersteds (Oe)
Delta_H = Delta_H * (1 /( 4 * np.pi *(10**(-3)))) #  (A/m)

# 0* Ms weber/m2 = 10**4 (4**Ms gauss)
M0 = 3000 / (4 * np.pi)   # Magnetizao (G)
M0 = ((10**(-4))/(mi0))* (4 * np.pi) * M0 # Magnetizao (weber/m2)

Mr = 2000 / (4 * np.pi)   # Magnetizao permanente (G)
Mr = ((10**(-4))/(mi0))* (4 * np.pi) * Mr # Magnetizao permanente (weber/m2)

Epsilon_p = 12.9  # Permissividade escalar
tan_delta = 0.0005 # loss tangent
Tc = 320  # Curie temperature

# Definindo a faixa de frequncias e calculando omega e Lambda
Frequencia = np.linspace(6 * 10**9, 12 * 10**9, 100)  # Faixa de 6 GHz a 12 GHz
omega = 2 * np.pi * Frequencia  # Frequncias angulares
Lambda = (C / Frequencia) * 10**3  # Convertendo Lambda para milmetros

# Configurando o grfico
plt.figure(figsize=(15, 10))

# Loop para diferentes valores de H0
for H0_value in H0_values:
    # Convertendo H0 para (A/m)
    H0 = H0_value * (1 / (4 * np.pi * (10**(-3))))  # (A/m)

    # Clculo de parmetros
    b =  mi0 * Gamma * Delta_H / (2 * omega)  # Constante de amortecimento adimensional
    omega_m = mi0 * Gamma * M0  # Frequncia de magnetizao
    omega_0 =  mi0 * Gamma * H0  # Frequncia de Larmor

    # Calculando os parmetros com dependncia de omega
    D = (((omega_0**2) - ((omega**2) * (1 + (b**2))))**2) + 4 * (omega_0**2) * (omega**2) * (b**2)
    X_1 = (omega_0 * omega_m * ((omega_0**2) - (omega**2) * (1 - (b**2)))) / D  # '
    X_2 = (omega * omega_m * b * ((omega_0**2) + (omega**2) * (1 + (b**2)))) / D  # ''
    omega_1 = (omega * omega_m * ((omega_0**2) - (omega**2) * (1 + (b**2)))) / D  # '
    omega_2 = (2 * (omega**2) * omega_0 * omega_m * b) / D  # ''

    # Definindo as variveis complexas mi_T e alfa
    mi_T = 1 + X_1 + X_2 * 1j  # _T = 1 + ' + i''
    alfa = omega_1 + omega_2 * 1j  #  = ' + i''

    # Separando as partes reais e imaginrias
    mi_T_real = mi_T.real
    mi_T_imaginaria = mi_T.imag
    alfa_real = alfa.real
    alfa_imaginaria = alfa.imag

    # Plotando as curvas para o valor de H0 atual
    plt.plot(Lambda, mi_T_real, label=f'Re{{_T}} para H0 = {H0_value}', linestyle='solid')
    plt.plot(Lambda, mi_T_imaginaria, label=f'Im{{_T}} para H0 = {H0_value}', linestyle='dotted')
    plt.plot(Lambda, alfa_real, label=f'Re{{}} para H0 = {H0_value}', linestyle='dashed')
    plt.plot(Lambda, alfa_imaginaria, label=f'Im{{}} para H0 = {H0_value}', linestyle='dashdot')

# Configuraes do grfico
plt.title('Dependncias de  dos parmetros de _r da Ferrita (Magnesium ferrite TT1-3000) com diferentes valores de H0', fontsize=20)
plt.xlabel(' [mm]', fontsize=20)
plt.ylabel('_p', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=12, frameon=True, facecolor='lightgrey', edgecolor='black')
plt.grid(True)
plt.tight_layout()
plt.show()
```
